# Churn Prediction
 
Since we know our best customers by segmentation and lifetime value prediction, we should also work hard on retaining them. That’s what makes Retention Rate is one of the most critical metrics.


Predicting which customer is going to churn is not a very easy task to perform and it depends on many variables. These variables highlight the behavior of the customer along with their demographic information. These variables come together to become very strong predictor of how satisfied or unsatisfied a particular customer is. The analysis of churn prediction can either be done at an individual level or at a cluster level. At a cluster level it comes down to segmenting people or users into buckets such that every user in a bucket has similar attributes.


In this article, we will use a Telco dataset and go over the following steps to develop a Churn Prediction model:

- Exploratory data analysis
- Feature engineering
- Investigating how the features affect Retention by using Logistic Regression
- Building a classification model with XGBoost

## 2. Feature Engineering

In this section, we are going to transform our raw features to extract more information from them. Our strategy is as follows:

1- Group the numerical columns by using clustering techniques
2- Apply Label Encoder to categorical features which are binary
3- Apply get_dummies() to categorical features which have multiple values


### 2.1 Numerical Columns

As we know from the EDA section, We have three numerical columns:

- Tenure
- Monthly Charges
- Total Charges

We are going to apply the following steps to create groups:
Using Elbow Method to identify the appropriate number of clusters, applying K-means logic to the selected column and change the naming and finally observe the profile of clusters

Let’s check how this works for Tenure in practice:

### 2.2 Cluster profiles:

```python
df_data.columns
```

```python
from sklearn.cluster import KMeans

def order_cluster(cluster_field_name, target_field_name, df, ascending):
    df_new = df.groupby(cluster_field_name)[target_field_name].mean().reset_index()
    df_new = df_new.sort_values(by=target_field_name,ascending=ascending).reset_index(drop=True)
    df_new['index'] = df_new.index
    df_final = pd.merge(df,df_new[[cluster_field_name,'index']], on=cluster_field_name)
    df_final = df_final.drop([cluster_field_name],axis=1)
    df_final = df_final.rename(columns={"index":cluster_field_name})
    return df_final

kmeans = KMeans(n_clusters=3)
kmeans.fit(df_data[['tenure']])
df_data['TenureCluster'] = kmeans.predict(df_data[['tenure']])


df_data = order_cluster('TenureCluster', 'tenure', df_data, False)
display(df_data.groupby('TenureCluster').tenure.describe())
```

```python
kmeans = KMeans(n_clusters=3)
kmeans.fit(df_data[['MonthlyCharges']])
df_data['MonthlyChargesCluster'] = kmeans.predict(df_data[['MonthlyCharges']])


df_data = order_cluster('MonthlyChargesCluster', 'MonthlyCharges', df_data, False)
display(df_data.groupby('MonthlyChargesCluster').MonthlyCharges.describe())
```

```python
kmeans = KMeans(n_clusters=3)

ind = (df_data['TotalCharges'] == ' ')
df_data.loc[ind, 'TotalCharges'] = 0
df_data['TotalCharges'] = df_data['TotalCharges'].astype('float')

kmeans.fit(df_data[['TotalCharges']])
df_data['TotalChargesCluster'] = kmeans.predict(df_data[['TotalCharges']])


df_data = order_cluster('TotalChargesCluster', 'TotalCharges', df_data, False)
display(df_data.groupby('TotalChargesCluster').TotalCharges.describe())
```

#### 2.2.1 One-Hot Transformation

```python
#import Label Encoder
from sklearn.preprocessing import LabelEncoder

columnNames = df_data.columns
columnNames = [i for i in columnNames if ('cluster' in i.lower())]
for i in columnNames:
    df_data[i] = df_data[i].astype(object) 

df_dataOneHot = df_data.copy()

le = LabelEncoder()
dummy_columns = [] #array for multiple value columns
for column in df_data.columns:
    if df_data[column].dtype == 'object' and column != 'customerID':
        if df_data[column].nunique() == 2:
            #apply Label Encoder for binary ones
            df_dataOneHot[column] = le.fit_transform(df_data[column]) 
        else:
            dummy_columns.append(column)
#apply get dummies for selected columns
  
    
df_dataOneHot = pd.get_dummies(data = df_dataOneHot,columns = dummy_columns)
```

```python
print(df_dataOneHot.shape, df_data.shape)
```

#### 2.2.2 Model Training

```python
from sklearn.model_selection import train_test_split
import xgboost as xgb
from sklearn.metrics import classification_report,confusion_matrix


#create feature set and labels
X = df_dataOneHot.drop(['Churn','customerID'],axis=1)
y = df_dataOneHot.Churn
#train and test split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.05, random_state=56)
#building the model & printing the score
xgb_model = xgb.XGBClassifier(max_depth=5, learning_rate=0.08, objective= 'binary:logistic',n_jobs=-1).fit(X_train, y_train)
print('Accuracy of XGB classifier on training set: {:.2f}'
       .format(xgb_model.score(X_train, y_train)))
print('Accuracy of XGB classifier on test set: {:.2f}'
       .format(xgb_model.score(X_test[X_train.columns], y_test)))
```

```python
from xgboost import plot_tree
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
import os
#install graphviz from https://graphviz.org/download/
os.environ["PATH"] += os.pathsep +'C:/Program Files/Graphviz/bin/'

y_pred = xgb_model.predict(X_test)
print(classification_report(y_test, y_pred))

fig, ax = plt.subplots(figsize=(100,100))
plot_tree(xgb_model, ax=ax)
```

```python
from xgboost import plot_importance
fig, ax = plt.subplots(figsize=(10,8))
plot_importance(xgb_model, ax=ax)
```

```python
df_dataOneHot['proba'] = xgb_model.predict_proba(df_dataOneHot[X_train.columns])[:,1]
df_dataOneHot[['customerID', 'proba']].head()
```
